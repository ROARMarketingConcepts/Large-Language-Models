## HuggingFace Transformers

```python
from transformers import pipeline
```

#### Sentiment Classifier

```python
sentiment_classifier = pipeline("sentiment-analysis")
```

```python
sentiment_classifier("I'm so excited to be learning about large language models")
```

#### Named Entity Recognition

```python
ner = pipeline("ner", model = "dslim/bert-base-NER")
```

```python
ner("Her name is Anna and she works in New York City for Morgan Stanley")
```

```python
zeroshot_classifier = pipeline("zero-shot-classification", model = "facebook/bart-large-mnli")
```

```python
sequence_to_classify = "one day I will see the world"
candidate_labels = ['travel', 'cooking', 'dancing']
```

```python
zeroshot_classifier(sequence_to_classify, candidate_labels)
```

## Pre-trained Tokenizers

**Note:** we need to specify the model being used when working with a tokenizer, as tokenizers are mostly unique to each model.

```python
from transformers import AutoTokenizer
```

```python
model = "bert-base-uncased"
```

```python
tokenizer = AutoTokenizer.from_pretrained(model)
```

```python
sentence = "I'm so excited to be learning about large language models"
```

token_type_ids: `0` for first sentence, `1` for second sentence, etc.

```python
input_ids = tokenizer(sentence)
print(input_ids)
```

```python
tokens = tokenizer.tokenize(sentence)
```

```python
print(tokens)
```

```python
token_ids = tokenizer.convert_tokens_to_ids(tokens)
```

```python
print(token_ids)
```

```python
decoded_ids = tokenizer.decode(token_ids)
print(decoded_ids)
```

**Classify token [CLS]** is a special token used in NLP and ML models, particularly those based on the Transformer architecture. It is a token that represents the entire input sequence or sentence and is placed at the beginning of the input.

```python
tokenizer.decode(101)
```

**Separate token [SEP]** is a special token used in natural language processing (NLP) and machine learning (ML) models to mark the separation between different segments of text. It plays a crucial role in tasks such as sentence classification, question answering, and text generation.

```python
tokenizer.decode(102)
```

```python
model2 = "xlnet-base-cased"
```

```python
tokenizer2 = AutoTokenizer.from_pretrained(model2)
```

```python
input_ids = tokenizer2(sentence)
```

```python
print(input_ids)
```

```python
tokens = tokenizer2.tokenize(sentence)
print(tokens)
```

## Huggingface and Pytorch/Tensorflow

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
```

```python
print(sentence)
print(input_ids)
```

```python
tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased-finetuned-sst-2-english")
```

Input ids returned as PyTorch tensors.

```python
input_ids_pt = tokenizer(sentence, return_tensors ="pt")
print(input_ids_pt)
```

```python
model = AutoModelForSequenceClassification.from_pretrained("distilbert-base-uncased-finetuned-sst-2-english")
```

```python
with torch.no_grad():
    logits = model(**input_ids_pt).logits

predicted_class_id = logits.argmax().item()
model.config.id2label[predicted_class_id]
```
